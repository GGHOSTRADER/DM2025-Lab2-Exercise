## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section

import pandas as pd

df1 = pd.read_csv(r"dm-lab-2-private-competition\data_identification.csv")
df1["post_id"] = df1["id"]
df1 = df1.drop(columns=["id"])
print(df1)
```

```python
df2 = pd.read_csv(r"dm-lab-2-private-competition\emotion.csv")
df2["post_id"] = df2["id"]
df2 = df2.drop(columns=["id"])
print(df2.emotion.unique())
print(df2)
```

```python
import json
import pandas as pd

type_text = []
post_id = []
text_list = []
hash_tags = []

with open(r"dm-lab-2-private-competition\final_posts.json", "r", encoding="utf-8") as f:
    data = json.load(f)

texts = [item["root"] for item in data]


for t in texts:
    type_text.append(t["_type"])
    post_id.append(t["_source"]["post"]["post_id"])
    text_list.append(t["_source"]["post"]["text"])
    hash_tags.append(t["_source"]["post"]["hashtags"])

df_clean = pd.DataFrame({
    "type": type_text,
    "post_id": post_id,
    "text": text_list,
    "hashtags": hash_tags
})



print(df_clean.head())
```

```python
print(len(df_clean.hashtags[64169]))
print(len(df_clean.hashtags[0]))
```

```python
df_merged_1 = df_clean.merge(df1, on="post_id", how="left")
df_merged_2 =  df_merged_1.merge(df2, on="post_id", how="left")

print(df_merged_1.shape)
print(df_merged_2.shape)
```

```python
print(df_merged_2["split"][df_merged_2["split"] == "test"].shape)
print(df_merged_2["split"][df_merged_2["split"] == "train"].shape)

# Checking how much information each column has (less counts == less information)
print(df_merged_2.describe())
# Type has 1 unique value only, will be dropped
df_merged_2 = df_merged_2.drop(columns=["type"])

# Noticed Hastags have empty lists
```

```python
# Cleaning hashtags column: replacing empty lists with 0
df_merged_2["hashtags_clean"] = df_merged_2["hashtags"].apply(lambda x: 0 if len(x) == 0 else x)
df_merged_2 = df_merged_2.drop(columns=["hashtags"])
# Creating data set without hashtags too
df_merged_2_nohastags = df_merged_2.copy().drop(columns=["hashtags_clean"])
print(df_merged_2.isnull().sum())
print(df_merged_2_nohastags.columns)
```

```python
df_merged_2[df_merged_2["hashtags_clean"] != 0][df_merged_2["split"] == "train"]
```

## 2. Feature Engineering Steps

```python
# Add hashtags of the texts that have into the sentence and create a longer setence with more information
def combine_text_and_tags(row):
    base = row["text"]
    tags = row["hashtags_clean"]
    if isinstance(tags, list):
        tag_str = " ".join(f"#{t}" for t in tags)
        return f"{base} {tag_str}"
    else:
        return base

df_merged_2["text_for_model"] = df_merged_2.apply(combine_text_and_tags, axis=1)

print(df_merged_2.head())
```

```python
# See how long are the sentences 
df_merged_2["length"]= df_merged_2["text_for_model"].apply(lambda x: len(x.split()))
# text length statistics
print(df_merged_2.length.describe())
```

```python
# Create training sets for text + hashtags  & text and no hashtags

# Text + Hashtags
df_train = df_merged_2[df_merged_2["split"] == "train"].copy()
df_test  = df_merged_2[df_merged_2["split"] == "test"].copy()   # no emotion labels
print(len(df_train), len(df_test))

# Text 
df_train_n = df_merged_2[df_merged_2["split"] == "train"].copy()
df_test_n  = df_merged_2[df_merged_2["split"] == "test"].copy()   # no emotion labels
print(len(df_train_n), len(df_test_n))

# Making sure Test has no labels
print(df_test.emotion.value_counts())
```

```python
# Map emotion strings to ints (text)
label2id_n = {label: i for i, label in enumerate((df_train_n["emotion"].unique()))}
id2label_n = {v: k for k, v in label2id_n.items()}
print(label2id_n)

df_train_n["label"] = df_train_n["emotion"].map(label2id_n)
print(df_train_n.head())
```

```python
# Map emotion strings to ints (text + hashtags)
label2id = {label: i for i, label in enumerate((df_train["emotion"].unique()))}
id2label = {v: k for k, v in label2id.items()}
print(label2id)

df_train["label"] = df_train["emotion"].map(label2id)
print(df_train.head())
```

```python
# Create Train + Validation set (text)
from sklearn.model_selection import train_test_split

train_df_n, val_df_n = train_test_split(
    df_train_n,
    test_size=0.2,
    stratify=df_train_n["label"],
    random_state=43,
)

print(len(train_df_n), len(val_df_n))
```

```python
# Create Train + Validation set (text + hashtags)

train_df, val_df = train_test_split(
    df_train,
    test_size=0.2,
    stratify=df_train["label"],
    random_state=42,
)

print(len(train_df), len(val_df))
```

# Deberta Non Linear Head (256 Hidden) +  No Smoothing 

```python
import numpy as np
from datasets import Dataset
from sklearn.metrics import accuracy_score, f1_score
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
import torch.nn as nn  # <<-- NEW

# ============================================================
# 1. HF Datasets from your pandas splits
# ============================================================

train_hf = Dataset.from_pandas(
    df_train[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)
val_hf = Dataset.from_pandas(
    val_df[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)

# ============================================================
# 2. DeBERTa v3 base tokenizer
# ============================================================

MODEL_NAME = "microsoft/deberta-v3-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 64  # enough for your data, much faster than 128

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN,
    )

train_hf = train_hf.map(tokenize_fn, batched=True)
val_hf   = val_hf.map(tokenize_fn, batched=True)

cols_to_keep = ["input_ids", "attention_mask", "label"]
train_hf = train_hf.remove_columns(
    [c for c in train_hf.column_names if c not in cols_to_keep]
)
val_hf   = val_hf.remove_columns(
    [c for c in val_hf.column_names if c not in cols_to_keep]
)

train_hf.set_format("torch")
val_hf.set_format("torch")

# ============================================================
# 3. Model + MLP head
# ============================================================

num_labels = len(label2id)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)

# ---- OVERRIDE DEFAULT LINEAR CLASSIFIER WITH 2-LAYER MLP ----
hidden_size = model.config.hidden_size

model.classifier = nn.Sequential(
    nn.Dropout(0.2),
    nn.Linear(hidden_size, 256),
    nn.GELU(),
    nn.Dropout(0.2),
    nn.Linear(256, num_labels),
)
# -------------------------------------------------------------

# ============================================================
# 4. TrainingArguments
# ============================================================

training_args = TrainingArguments(
    output_dir="./deberta_v3_emotion_hashtags",
    learning_rate=2e-5,
    per_device_train_batch_size=8,   # you can bump to 16 on the 4060 if it fits
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=100,
    save_total_limit=2,
)

# set these after construction to avoid the kwarg issue
training_args.evaluation_strategy = "epoch"
training_args.save_strategy = "epoch"
training_args.load_best_model_at_end = True
training_args.metric_for_best_model = "f1_macro"

print("=== Key training hyperparameters ===")
print("model_name:", MODEL_NAME)
print("output_dir:", training_args.output_dir)
print("num_train_epochs:", training_args.num_train_epochs)
print("learning_rate:", training_args.learning_rate)
print("train_batch_size:", training_args.per_device_train_batch_size)
print("eval_batch_size:", training_args.per_device_eval_batch_size)
print("weight_decay:", training_args.weight_decay)
print("logging_steps:", training_args.logging_steps)

# ============================================================
# 5. Metrics
# ============================================================

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro"),
    }

# ============================================================
# 6. Trainer + Train + Eval + Save
# ============================================================

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_hf,
    eval_dataset=val_hf,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

trainer.train()

metrics = trainer.evaluate()
print("Validation Accuracy:", metrics["eval_accuracy"])
print("Validation Macro F1:", metrics["eval_f1_macro"])
print("Validation metrics (DeBERTa v3 base + MLP head):", metrics)

trainer.save_model("./deberta_v3_emotion_hashtags/best_model")
```

```python
from transformers import Trainer
import torch

class WeightedTrainer(Trainer):
    def __init__(self, class_weights, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(
        self,
        model,
        inputs,
        return_outputs: bool = False,
        num_items_in_batch: int | None = None,  # <- accept extra arg, ignore it
    ):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        weights = self.class_weights.to(logits.device)
        loss_fct = torch.nn.CrossEntropyLoss(weight=weights)
        loss = loss_fct(
            logits.view(-1, model.config.num_labels),
            labels.view(-1),
        )

        if return_outputs:
            return loss, outputs
        return loss
```
