## 2. Feature Engineering Steps

```python
# Add hashtags of the texts that have into the sentence and create a longer setence with more information
def combine_text_and_tags(row):
    base = row["text"]
    tags = row["hashtags_clean"]
    if isinstance(tags, list):
        tag_str = " ".join(f"#{t}" for t in tags)
        return f"{base} {tag_str}"
    else:
        return base

df_merged_2["text_for_model"] = df_merged_2.apply(combine_text_and_tags, axis=1)

print(df_merged_2.head())
```

```python
# See how long are the sentences 
df_merged_2["length"]= df_merged_2["text_for_model"].apply(lambda x: len(x.split()))
# text length statistics
print(df_merged_2.length.describe())

```

```python
# Create training sets for text + hashtags  & text and no hashtags

# Text + Hashtags
df_train = df_merged_2[df_merged_2["split"] == "train"].copy()
df_test  = df_merged_2[df_merged_2["split"] == "test"].copy()   # no emotion labels
print(len(df_train), len(df_test))

# Text 
df_train_n = df_merged_2[df_merged_2["split"] == "train"].copy()
df_test_n  = df_merged_2[df_merged_2["split"] == "test"].copy()   # no emotion labels
print(len(df_train_n), len(df_test_n))

# Making sure Test has no labels
print(df_test.emotion.value_counts())
```

```python
# Map emotion strings to ints (text)
label2id_n = {label: i for i, label in enumerate((df_train_n["emotion"].unique()))}
id2label_n = {v: k for k, v in label2id_n.items()}
print(label2id_n)
df_train_n["label"] = df_train_n["emotion"].map(label2id_n)
print(df_train_n.head())
```

```python
# Map emotion strings to ints (text + hashtags)
label2id = {label: i for i, label in enumerate((df_train["emotion"].unique()))}
id2label = {v: k for k, v in label2id.items()}
print(label2id)

df_train["label"] = df_train["emotion"].map(label2id)
print(df_train.head())
```

```python
# Create Train + Validation set (text)
from sklearn.model_selection import train_test_split


train_df_n, val_df_n = train_test_split(
    df_train_n,
    test_size=0.2,
    stratify=df_train_n["label"],
    random_state=43,
)

print(len(train_df_n), len(val_df_n))
```

# 3. Model Implementation Steps

## Deberta Linear Head + hashtags + FASTER Learning (Train/Val) --Ready

```python
import numpy as np
from datasets import Dataset
from sklearn.metrics import accuracy_score, f1_score
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)

# ------------------------------------------------------
# 1. HF Datasets from your pandas splits

train_hf = Dataset.from_pandas(
    df_train[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)
val_hf = Dataset.from_pandas(
    val_df[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)
# ------------------------------------------------------
# 2. DeBERTa v3 base tokenizer

MODEL_NAME = "microsoft/deberta-v3-base" # <---- important, it loads the model and if changed can use fine tunned one
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 64  # much faster than 128

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN,
    )

train_hf = train_hf.map(tokenize_fn, batched=True)
val_hf   = val_hf.map(tokenize_fn, batched=True)

cols_to_keep = ["input_ids", "attention_mask", "label"]
train_hf = train_hf.remove_columns([c for c in train_hf.column_names if c not in cols_to_keep])
val_hf   = val_hf.remove_columns([c for c in val_hf.column_names if c not in cols_to_keep])

train_hf.set_format("torch")
val_hf.set_format("torch")

# ------------------------------------------------------
# 3. Model

num_labels = len(label2id)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)
# ------------------------------------------------------
# 4. TrainingArguments

training_args = TrainingArguments(
    output_dir="./deberta_v3_emotion_hashtags",
    learning_rate=2e-5,
    per_device_train_batch_size=8,   # safer for v3-base on my hardware
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=100,
    save_total_limit=2,
)


training_args.evaluation_strategy = "epoch"
training_args.save_strategy = "epoch"
training_args.load_best_model_at_end = True
training_args.metric_for_best_model = "f1_macro"

print("=== Key training hyperparameters ===")
print("model_name:", MODEL_NAME)
print("output_dir:", training_args.output_dir)
print("num_train_epochs:", training_args.num_train_epochs)
print("learning_rate:", training_args.learning_rate)
print("train_batch_size:", training_args.per_device_train_batch_size)
print("eval_batch_size:", training_args.per_device_eval_batch_size)
print("weight_decay:", training_args.weight_decay)
print("logging_steps:", training_args.logging_steps)

# ------------------------------------------------------
# 5. Metrics

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro"),
    }

# ------------------------------------------------------
# 6. Trainer + Train + Eval + Save

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_hf,
    eval_dataset=val_hf,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

trainer.train()

metrics = trainer.evaluate()
print("Validation metrics (DeBERTa v3 base):", metrics)

trainer.save_model("./deberta_v3_emotion_hashtags/best_model")
```

```python
# Logit generator from saved models in my DIR, 

import numpy as np
import pandas as pd
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)

# ------------------------------------------------------
# 0. Basic config

DEVICE    = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_DIR = "./deberta_v3_test_new_code/final_model_dec_1"  # change model 

ID_COL   = "post_id"
TEXT_COL = "text_for_model"
MAX_LEN  = 64

# ------------------------------------------------------
# 1. Load tokenizer + model from checkpoint

tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_DIR)
model.to(DEVICE)
model.eval()

# grab label mapping from the checkpoint 
id2label = model.config.id2label  
label2id = model.config.label2id

print("Loaded model from:", MODEL_DIR)
print("id2label:", id2label)

# ------------------------------------------------------
# 2. Build HF Dataset from df_test


# Change here from df_train to df_test to buil different logits
df_test_for_hf = df_train[[ID_COL, TEXT_COL]].rename(
    columns={TEXT_COL: "text", ID_COL: "id"}
)

test_hf = Dataset.from_pandas(df_test_for_hf)

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN,
    )

test_hf = test_hf.map(tokenize_fn, batched=True)


ids = test_hf["id"]

# keep only model inputs as columns
cols_to_keep = ["input_ids", "attention_mask"]
test_hf = test_hf.remove_columns(
    [c for c in test_hf.column_names if c not in cols_to_keep]
)
test_hf.set_format("torch")

# ------------------------------------------------------
# 3. Trainer for inference only

inference_args = TrainingArguments(
    output_dir="./deberta_idk",
    per_device_eval_batch_size=32,
    dataloader_drop_last=False,
    do_train=False,
    do_eval=False,
    do_predict=True,
    logging_steps=50,
)

trainer = Trainer(
    model=model,
    args=inference_args,
    tokenizer=tokenizer,
)

# ------------------------------------------------------
# 4. Predict: get logits

pred_output = trainer.predict(test_hf)
logits = pred_output.predictions  # shape [N, num_labels]
print("Logits shape:", logits.shape)

#  Save logits 
np.save("deberta_logits_1dec_train.npy", logits)

# DataFrame with one column per logit

def _sorted_label_keys(id2label_dict):
    keys = list(id2label_dict.keys())
    # convert numeric-like strings to ints for sorting
    def key_fn(k):
        if isinstance(k, int):
            return k
        if isinstance(k, str) and k.isdigit():
            return int(k)
        return k
    return sorted(keys, key=key_fn)

sorted_keys = _sorted_label_keys(id2label)
logit_cols = [f"logit_{id2label[k]}" for k in sorted_keys]

logits_df = pd.DataFrame(logits, columns=logit_cols)
logits_df.insert(0, "id", np.array(ids))
logits_df.to_csv("deberta_logits_1dec_train.csv", index=False)

print("Saved logits to:")
print("  - deberta_hashtags_bestmodel_test_logits.npy")
print("  - deberta_hashtags_bestmodel_test_logits.csv")

# ------------------------------------------------------
# 5. Turn logits into hard labels + Kaggle submission format


pred_ids = logits.argmax(axis=-1)  # [N]

pred_labels = []
for i in pred_ids:
    # try both string and int keys just in case
    if str(i) in id2label:
        pred_labels.append(id2label[str(i)])
    else:
        pred_labels.append(id2label[int(i)])

submission = pd.DataFrame({
    "id": np.array(ids),  
    "emotion": pred_labels,
})

submission_path = "submission_test_inference_deberta_train.csv"
submission.to_csv(submission_path, index=False)
print("Saved submission to:", submission_path)
```

## Feature engineering

```python
import numpy as np
import pandas as pd
import lightgbm as lgb

from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack, csr_matrix

import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# ------------------------------------------------------
# 1. Simple features

def add_handcrafted_features(df, text_col="text_for_model"):
    """
    Hand-crafted text features for LightGBM, combining:
      - Original features (len_chars, len_words, caps_ratio, etc.)
      - Extra features (unique words, digits, non-ascii, etc.)
    """
    df = df.copy()
    s = df[text_col].fillna("")


    # Basic length features
    df["len_chars"] = s.str.len()
    df["len_words"] = s.str.split().str.len()

    # Punctuation / intensity
    df["num_exclam"] = s.str.count("!")
    df["num_question"] = s.str.count(r"\?")

    # Uppercase ratio (shouting) 
    def caps_ratio_fn(x):
        alpha = sum(c.isalpha() for c in x)
        if alpha == 0:
            return 0.0
        return sum(c.isupper() for c in x) / alpha

    df["caps_ratio"] = s.apply(caps_ratio_fn)

    # Structural: hashtags, mentions, urls 
    df["num_hashtags"] = s.str.count("#")
    df["num_mentions"] = s.str.count("@")
    df["num_urls"] = s.str.count(r"http")


    #  Word-level stats 
    len_words_safe = df["len_words"].replace(0, np.nan)

    # Unique words
    df["n_unique_words"] = s.str.split().apply(
        lambda x: len(set(x)) if isinstance(x, list) and len(x) > 0 else 0
    )

    # Ratios
    df["ratio_unique_words"] = df["n_unique_words"] / len_words_safe
    df["chars_per_word"] = df["len_chars"] / len_words_safe

    df[["ratio_unique_words", "chars_per_word"]] = df[
        ["ratio_unique_words", "chars_per_word"]
    ].fillna(0.0)

    df["ratio_unique_words"] = df["ratio_unique_words"].clip(0, 1)

    #  Extra punctuation 
    df["n_period"] = s.str.count(r"\.")
    df["n_comma"] = s.str.count(",")

    #  Capitalization (raw count) 
    df["n_caps"] = s.apply(lambda x: sum(1 for ch in x if ch.isupper()))

    #  Digits / non-ASCII 
    df["n_digits"] = s.str.count(r"\d")
    df["n_non_ascii"] = s.apply(lambda x: sum(ord(ch) > 127 for ch in x))

    #  Whitespace patterns 
    df["n_newlines"] = s.str.count(r"\n")
    df["n_tabs"] = s.str.count(r"\t")

    return df

```

```python

# ------------------------------------------------------
# 2. Sentiment feature function Using ROBERTA

SENT_MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment-latest"

try:
    sent_tokenizer = AutoTokenizer.from_pretrained(
        SENT_MODEL_NAME,
        local_files_only=False,   # set True if you know it's cached
    )
    sent_model = AutoModelForSequenceClassification.from_pretrained(
        SENT_MODEL_NAME,
        local_files_only=False,
    )
    print("Loaded sentiment model.")
except Exception as e:
    print("Failed to load sentiment model:", e)
    sent_tokenizer = None
    sent_model = None

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
if sent_model is not None:
    sent_model.to(device)
    sent_model.eval()


def get_sentiment_features(
    df,
    text_col="text_for_model",
    batch_size=128,
    max_len=48,
):
    
    df = df.copy()

    # Fallback if model not available
    if sent_model is None or sent_tokenizer is None:
        print("Sentiment model unavailable; using dummy neutral features.")
        return pd.DataFrame(
            {
                "sent_negative": np.zeros(len(df), dtype=float),
                "sent_neutral": np.ones(len(df), dtype=float),
                "sent_positive": np.zeros(len(df), dtype=float),
            },
            index=df.index,
        )

    texts = df[text_col].fillna("").tolist()
    if len(texts) == 0:
        return pd.DataFrame(
            {
                "sent_negative": [],
                "sent_neutral": [],
                "sent_positive": [],
            },
            index=df.index,
        )

    enc = sent_tokenizer(
        texts,
        padding=True,
        truncation=True,
        max_length=max_len,
        return_tensors="pt",
    )

    input_ids = enc["input_ids"]
    attention_mask = enc["attention_mask"]

    n = input_ids.size(0)
    all_probs = []

    with torch.no_grad():
        for start in range(0, n, batch_size):
            end = min(start + batch_size, n)
            ids_batch = input_ids[start:end].to(device)
            mask_batch = attention_mask[start:end].to(device)

            outputs = sent_model(input_ids=ids_batch, attention_mask=mask_batch)
            logits = outputs.logits  # (B, 3)
            probs = F.softmax(logits, dim=-1).cpu().numpy()

            all_probs.append(probs)

            if start == 0 or (start // batch_size) % 10 == 0:
                print(f"Sentiment: processed {end}/{n}")

    all_probs = np.vstack(all_probs)  # (N, 3)

    id2label = sent_model.config.id2label
    num_labels = all_probs.shape[1]

    cols = {}
    for i in range(num_labels):
        label_name = id2label.get(i, f"label_{i}").lower()
        cols[f"sent_{label_name}"] = all_probs[:, i]

    sent_df = pd.DataFrame(cols, index=df.index)

    # Ensure all sentiment columns exist
    for col in ["sent_negative", "sent_neutral", "sent_positive"]:
        if col not in sent_df.columns:
            sent_df[col] = 0.0

    return sent_df
```

```python
# ------------------------------------------------------
# 3. Build handcrafted + sentiment features

HAS_VAL = "val_df" in globals() and val_df is not None

df_train_feats = add_handcrafted_features(df_train, text_col="text_for_model")
train_sent = get_sentiment_features(df_train_feats, text_col="text_for_model")
df_train_feats = pd.concat([df_train_feats, train_sent], axis=1)

if HAS_VAL:
    df_val_feats = add_handcrafted_features(val_df, text_col="text_for_model")
    val_sent = get_sentiment_features(df_val_feats, text_col="text_for_model")
    df_val_feats = pd.concat([df_val_feats, val_sent], axis=1)
else:
    df_val_feats = None

df_test_feats = add_handcrafted_features(df_test, text_col="text_for_model")
test_sent = get_sentiment_features(df_test_feats, text_col="text_for_model")
df_test_feats = pd.concat([df_test_feats, test_sent], axis=1)

print(
    "Train feature columns added:",
    [c for c in df_train_feats.columns if c not in df_train.columns],
)


# add external logits DeBERTweet and  DeBERTa as features

def add_logits_features(
    base_df: pd.DataFrame,
    logits_df: pd.DataFrame,
    model_tag: str,
    join_col_base: str = "id",
    join_col_logits: str = "id",
):
    
    #Merge logits_df into base_df on id, prefix columns with model_tag.
    #Returns: merged_df, list_of_new_cols
    
    base_df = base_df.copy()
    logits_df = logits_df.copy()

    if join_col_base not in base_df.columns and "post_id" in base_df.columns:
        join_col_base = "post_id"

    if join_col_logits not in logits_df.columns:
        raise ValueError(f"logits_df missing join column '{join_col_logits}'")

    # Identify logit columns (all numeric except id/post_id)
    candidate_cols = [
        c for c in logits_df.columns
        if c not in ["id", "post_id"] and pd.api.types.is_numeric_dtype(logits_df[c])
    ]
    if not candidate_cols:
        raise ValueError(f"No numeric logit columns found in logits_df for {model_tag}")

    
    rename_map = {}
    for i, c in enumerate(candidate_cols):
        new_name = f"{model_tag}_logit_{i}"
        rename_map[c] = new_name

    logits_df = logits_df[[join_col_logits] + candidate_cols].rename(
        columns={join_col_logits: join_col_base, **rename_map}
    )

    merged = base_df.merge(logits_df, on=join_col_base, how="left")

    new_cols = list(rename_map.values())
    merged[new_cols] = merged[new_cols].astype(float)

    return merged, new_cols

# ------------------------------------------------------
#  TRAIN: add DeBERTweet + DeBERTa logits 

df_train_feats, train_debertweet_cols = add_logits_features(
    df_train_feats,
    df_train_debertweet_slow,
    model_tag="debertweet",
    join_col_base="id",
    join_col_logits="id",
)

df_train_feats, train_deberta_cols = add_logits_features(
    df_train_feats,
    df_train_deberta_slow,
    model_tag="deberta",
    join_col_base="id",
    join_col_logits="id",
)

logit_feature_cols = train_debertweet_cols + train_deberta_cols
print("Added train logit feature columns:", logit_feature_cols)

# ------------------------------------------------------
# 2) VAL: add same columns as zeros (no logits) 

if HAS_VAL and df_val_feats is not None:
    for col in logit_feature_cols:
        if col not in df_val_feats.columns:
            df_val_feats[col] = 0.0

# ------------------------------------------------------
# 3) TEST: add DeBERTweet + DeBERTa logits 

df_test_feats, test_debertweet_cols = add_logits_features(
    df_test_feats,
    df_test_debertweet_slow,
    model_tag="debertweet",
    join_col_base="id",
    join_col_logits="id",
)

df_test_feats, test_deberta_cols = add_logits_features(
    df_test_feats,
    df_test_deberta_slow,
    model_tag="deberta",
    join_col_base="id",
    join_col_logits="id",
)

assert set(logit_feature_cols) == set(test_debertweet_cols + test_deberta_cols), \
    "Train/test logit column mismatch"

print("Train shape after adding logits:", df_train_feats.shape)
if HAS_VAL:
    print("Val shape after adding logits:", df_val_feats.shape)
print("Test shape after adding logits:", df_test_feats.shape)
```

```python

# ------------------------------------------------------
# 4. TF-IDF features (unigrams + bigrams)

tfidf = TfidfVectorizer(
    max_features=20000,
    ngram_range=(1, 2),
    min_df=2,
    max_df=0.9,
)

# Fit on train text; transform val/test
tfidf_train = tfidf.fit_transform(df_train["text_for_model"].fillna(""))
tfidf_val = tfidf.transform(val_df["text_for_model"].fillna("")) if HAS_VAL else None
tfidf_test = tfidf.transform(df_test["text_for_model"].fillna(""))

print("TF-IDF shapes:", tfidf_train.shape,
      tfidf_val.shape if tfidf_val is not None else None,
      tfidf_test.shape)
```

## Model Building Steps

```python

# ------------------------------------------------------
# LightGBM with numeric + TF-IDF


def train_lightgbm_on_features(
    df_train_feats,
    tfidf_train,
    df_val_feats=None,
    tfidf_val=None,
    label_col="label",
    drop_cols=("text_for_model", "id", "post_id"),
    label2id=None,
    num_classes=None,
):
    assert isinstance(df_train_feats, pd.DataFrame), "df_train_feats must be a DataFrame"
    if df_val_feats is not None:
        assert isinstance(df_val_feats, pd.DataFrame), "df_val_feats must be a DataFrame or None"

    # Targets
    y_train = df_train_feats[label_col].values
    y_val = None
    if df_val_feats is not None and label_col in df_val_feats.columns:
        y_val = df_val_feats[label_col].values

    # Num classes
    if num_classes is None:
        if label2id is not None:
            num_classes = len(label2id)
        else:
            num_classes = len(np.unique(y_train))

    # Numeric feature selection
    numeric_cols = [
        c for c in df_train_feats.columns
        if pd.api.types.is_numeric_dtype(df_train_feats[c])
    ]
    feature_cols = [
        c for c in numeric_cols
        if c != label_col and c not in drop_cols
    ]

    # Numeric to sparse
    X_train_num = csr_matrix(df_train_feats[feature_cols].values)
    X_train = hstack([X_train_num, tfidf_train])  # sparse matrix

    if df_val_feats is not None and tfidf_val is not None:
        X_val_num = csr_matrix(df_val_feats[feature_cols].values)
        X_val = hstack([X_val_num, tfidf_val])
    else:
        X_val = None

    print("Using numeric features:", feature_cols)
    print("X_train (numeric + tfidf) shape:", X_train.shape)
    if X_val is not None:
        print("X_val (numeric + tfidf) shape:", X_val.shape)
    else:
        print("No validation set provided; training without early stopping.")


    # LightGBM model
    lgbm = lgb.LGBMClassifier(
        objective="multiclass",
        num_class=num_classes,
        learning_rate=0.05,
        n_estimators=2000,
        max_depth=-1,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        n_jobs=-1,
    )

    # Fit
    if X_val is not None and y_val is not None:
        lgbm.fit(
            X_train,
            y_train,
            eval_set=[(X_val, y_val)],
            eval_metric="multi_logloss",
            early_stopping_rounds=50,
            verbose=50,
        )

        # Eval
        y_val_pred = lgbm.predict(X_val)
        val_acc = accuracy_score(y_val, y_val_pred)
        val_f1_macro = f1_score(y_val, y_val_pred, average="macro")

        print("\n=== LightGBM (numeric + tfidf, with val) ===")
        print("Validation Accuracy:", val_acc)
        print("Validation Macro F1:", val_f1_macro)

        if label2id is not None:
            id2label_local = {v: k for k, v in label2id.items()}
            target_names = [id2label_local.get(i, str(i)) for i in range(num_classes)]
        else:
            target_names = [str(i) for i in range(num_classes)]

        print("\n=== Classification report (per class) ===")
        print(classification_report(y_val, y_val_pred, target_names=target_names))
    else:
        lgbm.fit(X_train, y_train)
        print("\n=== LightGBM (numeric + tfidf, no val set) ===")
        print("Trained on full train set without early stopping.")

    return lgbm, feature_cols
# ------------------------------------------------------
# 6. Train model


if HAS_VAL:
    model_lgbm, feature_cols = train_lightgbm_on_features(
        df_train_feats,
        tfidf_train,
        df_val_feats=df_val_feats,
        tfidf_val=tfidf_val,
        label_col="label",
        drop_cols=("text_for_model", "id", "post_id"),
        label2id=label2id,
    )
else:
    model_lgbm, feature_cols = train_lightgbm_on_features(
        df_train_feats,
        tfidf_train,
        df_val_feats=None,
        tfidf_val=None,
        label_col="label",
        drop_cols=("text_for_model", "id", "post_id"),
        label2id=label2id,
    )

# ------------------------------------------------------
# 7. Predict on test and build Kaggle submission


# Build numeric + tfidf for test
X_test_num = csr_matrix(df_test_feats[feature_cols].values)
X_test = hstack([X_test_num, tfidf_test])

test_pred = model_lgbm.predict(X_test)  # integer labels

# Map back to emotion strings
id2label_local = {v: k for k, v in label2id.items()}
test_emotion = [id2label_local[int(i)] for i in test_pred]


if "post_id" in df_test_feats.columns:
    test_ids = df_test_feats["post_id"].values
elif "post_id" in df_test.columns:
    test_ids = df_test["post_id"].values
else:
    raise ValueError("No 'post_id' column found in df_test_feats or df_test.")

submission = pd.DataFrame({
    "id": test_ids,         
    "emotion": test_emotion,
})

submission_path = "submission_lightgbm_feats_sentiment_tfidf_logits.csv"
submission.to_csv(submission_path, index=False)
print(f"\nSaved Kaggle submission to: {submission_path}")
```

# HARD VOTING FINAL CSV

```python
import pandas as pd
from collections import Counter

# ------------------------------------------------------
# 1. Path output CSVs

output_path         = "submission_hard_vote_models_2.csv"

# ------------------------------------------------------
# 2. Load CSVs

df_deberta_no_val = pd.read_csv("YES_submission_bertweet_large_no_nov2_new_code_normal.csv")
df_lightgbm       = pd.read_csv("YES_submission_deberta_v3_no_val_slow.csv")
df_bertweet_slow  = pd.read_csv("YES_submission_lightgbm_feats_sentiment_tfidf_logits.csv")
df_bertweet       = pd.read_csv("YES_submission_deberta_v3_no_val_slow_nov2.csv")
df_deberta        = pd.read_csv("YES_submission_deberta_v3.csv")

# ------------------------------------------------------
# 3. Sort by id to ensure alignment

dfs = [df_deberta_no_val, df_lightgbm, df_bertweet_slow, df_bertweet, df_deberta]
dfs = [df.sort_values("id").reset_index(drop=True) for df in dfs]

#  IDs must match in all files
base_ids = dfs[0]["id"]
for i, df in enumerate(dfs[1:], start=2):
    if not base_ids.equals(df["id"]):
        raise ValueError(f"ID mismatch between base file and file #{i}")

# ------------------------------------------------------
# 4. Hard voting with 5 models
# Rules
# 1) Majority label wins
# 2) If all 5 different → deberta_no_val_slow will be selected if all are different
# 
e_no_val   = dfs[0]["emotion"]
e_lgbm     = dfs[1]["emotion"]
e_btw_slow = dfs[2]["emotion"]
e_btw      = dfs[3]["emotion"]
e_deberta  = dfs[4]["emotion"]

final_emotions = []

for e1, e2, e3, e4, e5 in zip(e_no_val, e_lgbm, e_btw_slow, e_btw, e_deberta):
    votes = [e1, e2, e3, e4, e5]
    counts = Counter(votes)
    label, count = counts.most_common(1)[0]

    if count == 1:
        final_emotions.append(e1)
    else:
        final_emotions.append(label)

# ------------------------------------------------------
# 5. Build final DataFrame & save

df_final = pd.DataFrame({
    "id": base_ids,
    "emotion": final_emotions,
})

df_final.to_csv(output_path, index=False)
print(f"Saved 5-model hard-vote submission to: {output_path}")
```
