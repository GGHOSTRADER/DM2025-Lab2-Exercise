```python
import json
import pandas as pd

type_text = []
post_id = []
text_list = []
hash_tags = []

with open(r"dm-lab-2-private-competition\final_posts.json", "r", encoding="utf-8") as f:
    data = json.load(f)

texts = [item["root"] for item in data]


for t in texts:
    type_text.append(t["_type"])
    post_id.append(t["_source"]["post"]["post_id"])
    text_list.append(t["_source"]["post"]["text"])
    hash_tags.append(t["_source"]["post"]["hashtags"])

df_clean = pd.DataFrame({
    "type": type_text,
    "post_id": post_id,
    "text": text_list,
    "hashtags": hash_tags
})



print(df_clean.head())
```

## 2. Feature Engineering Steps

```python
# Add hashtags of the texts that have into the sentence and create a longer setence with more information
def combine_text_and_tags(row):
    base = row["text"]
    tags = row["hashtags_clean"]
    if isinstance(tags, list):
        tag_str = " ".join(f"#{t}" for t in tags)
        return f"{base} {tag_str}"
    else:
        return base

df_merged_2["text_for_model"] = df_merged_2.apply(combine_text_and_tags, axis=1)

print(df_merged_2.head())
```

# Deberta Non Linear Head (256 Hidden) +  No Smoothing 

```python
import numpy as np
from datasets import Dataset
from sklearn.metrics import accuracy_score, f1_score
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
import torch.nn as nn  # <<-- NEW

# ============================================================
# 1. HF Datasets from your pandas splits
# ============================================================

train_hf = Dataset.from_pandas(
    df_train[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)
val_hf = Dataset.from_pandas(
    val_df[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)

# ============================================================
# 2. DeBERTa v3 base tokenizer
# ============================================================

MODEL_NAME = "microsoft/deberta-v3-base"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 64  # enough for your data, much faster than 128

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=MAX_LEN,
    )

train_hf = train_hf.map(tokenize_fn, batched=True)
val_hf   = val_hf.map(tokenize_fn, batched=True)

cols_to_keep = ["input_ids", "attention_mask", "label"]
train_hf = train_hf.remove_columns(
    [c for c in train_hf.column_names if c not in cols_to_keep]
)
val_hf   = val_hf.remove_columns(
    [c for c in val_hf.column_names if c not in cols_to_keep]
)

train_hf.set_format("torch")
val_hf.set_format("torch")

# ============================================================
# 3. Model + MLP head
# ============================================================

num_labels = len(label2id)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)

# ---- OVERRIDE DEFAULT LINEAR CLASSIFIER WITH 2-LAYER MLP ----
hidden_size = model.config.hidden_size

model.classifier = nn.Sequential(
    nn.Dropout(0.2),
    nn.Linear(hidden_size, 256),
    nn.GELU(),
    nn.Dropout(0.2),
    nn.Linear(256, num_labels),
)
# -------------------------------------------------------------

# ============================================================
# 4. TrainingArguments
# ============================================================

training_args = TrainingArguments(
    output_dir="./deberta_v3_emotion_hashtags",
    learning_rate=2e-5,
    per_device_train_batch_size=8,   # you can bump to 16 on the 4060 if it fits
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=100,
    save_total_limit=2,
)

# set these after construction to avoid the kwarg issue
training_args.evaluation_strategy = "epoch"
training_args.save_strategy = "epoch"
training_args.load_best_model_at_end = True
training_args.metric_for_best_model = "f1_macro"

print("=== Key training hyperparameters ===")
print("model_name:", MODEL_NAME)
print("output_dir:", training_args.output_dir)
print("num_train_epochs:", training_args.num_train_epochs)
print("learning_rate:", training_args.learning_rate)
print("train_batch_size:", training_args.per_device_train_batch_size)
print("eval_batch_size:", training_args.per_device_eval_batch_size)
print("weight_decay:", training_args.weight_decay)
print("logging_steps:", training_args.logging_steps)

# ============================================================
# 5. Metrics
# ============================================================

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro"),
    }

# ============================================================
# 6. Trainer + Train + Eval + Save
# ============================================================

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_hf,
    eval_dataset=val_hf,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

trainer.train()

metrics = trainer.evaluate()
print("Validation Accuracy:", metrics["eval_accuracy"])
print("Validation Macro F1:", metrics["eval_f1_macro"])
print("Validation metrics (DeBERTa v3 base + MLP head):", metrics)

trainer.save_model("./deberta_v3_emotion_hashtags/best_model")
```

```python
from transformers import Trainer
import torch

class WeightedTrainer(Trainer):
    def __init__(self, class_weights, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(
        self,
        model,
        inputs,
        return_outputs: bool = False,
        num_items_in_batch: int | None = None,  # <- accept extra arg, ignore it
    ):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        weights = self.class_weights.to(logits.device)
        loss_fct = torch.nn.CrossEntropyLoss(weight=weights)
        loss = loss_fct(
            logits.view(-1, model.config.num_labels),
            labels.view(-1),
        )

        if return_outputs:
            return loss, outputs
        return loss
```

```python
import numpy as np
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
)
from sklearn.metrics import accuracy_score, f1_score

# ============================================================
# 1. Build HF Datasets from train/val DataFrames
#    Assumes train_df and val_df have: "text_for_model" and "label"
# ============================================================

train_hf = Dataset.from_pandas(
    train_df[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)
val_hf = Dataset.from_pandas(
    val_df[["text_for_model", "label"]].rename(columns={"text_for_model": "text"})
)

MODEL_NAME = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def tokenize_fn(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=70,
    )

train_hf = train_hf.map(tokenize_fn, batched=True)
val_hf   = val_hf.map(tokenize_fn, batched=True)

cols = ["input_ids", "attention_mask", "label"]
train_hf = train_hf.remove_columns([c for c in train_hf.column_names if c not in cols])
val_hf   = val_hf.remove_columns([c for c in val_hf.column_names if c not in cols])

train_hf.set_format("torch")
val_hf.set_format("torch")

# ============================================================
# 2. Model
# ============================================================

num_labels = len(label2id)

model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)

# ============================================================
# 3. TrainingArguments (minimal to avoid your kwarg issues)
# ============================================================

training_args = TrainingArguments(
    output_dir="./distilbert_emotion_hashtags",
    learning_rate=2e-5,
    per_device_train_batch_size=16,   # drop to 8/4 if CUDA OOM
    per_device_eval_batch_size=32,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=100,
    save_total_limit=2,
)

# ============================================================
# 4. Metrics function
# ============================================================

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1_macro": f1_score(labels, preds, average="macro"),
    }

# ============================================================
# 5. Trainer + Train + Eval + Save
# ============================================================

trainer = WeightedTrainer(
    class_weights=class_weights,   # <- new
    model=model,
    args=training_args,
    train_dataset=train_hf,
    eval_dataset=val_hf,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)


trainer.train()

metrics = trainer.evaluate()
print("Validation metrics:", metrics)

save_path = "./distilbert_emotion_hashtags/best_model"
trainer.save_model(save_path)
tokenizer.save_pretrained(save_path)
print(f"Model saved to {save_path}")
```

```python
import re
import numpy as np
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# ============================================================
# 0. RoBERTa sentiment model (feature extractor only)
# ============================================================

SENT_MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment-latest"

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

sent_tokenizer = AutoTokenizer.from_pretrained(SENT_MODEL_NAME)
sent_model = AutoModelForSequenceClassification.from_pretrained(SENT_MODEL_NAME)
sent_model.to(device)
sent_model.eval()

def compute_roberta_sentiment(texts, batch_size=32, max_len=128):
    """
    Compute sentiment probs (neg, neu, pos) using Twitter RoBERTa.
    Returns an (N, 3) numpy array.
    """
    all_scores = []

    for start in range(0, len(texts), batch_size):
        batch = texts[start:start + batch_size]
        enc = sent_tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=max_len,
            return_tensors="pt",
        )
        enc = {k: v.to(device) for k, v in enc.items()}

        with torch.no_grad():
            outputs = sent_model(**enc)
            probs = outputs.logits.softmax(dim=-1).cpu().numpy()
            # cardiffnlp/twitter-roberta-base-sentiment-latest labels:
            # 0 = negative, 1 = neutral, 2 = positive
            all_scores.append(probs)

    return np.vstack(all_scores) if all_scores else np.zeros((0, 3), dtype=float)

# ============================================================
# 1. Handcrafted feature function + sentiment
# ============================================================

def add_handcrafted_features(df, text_col="text_for_model"):
    """
    Add simple, fast, language-agnostic text features + RoBERTa sentiment.
    Returns a copy of df with new feature columns appended.
    """
    df = df.copy()
    s = df[text_col].fillna("")

    # ---------- Basic length features ----------
    df["n_chars"] = s.str.len()
    df["n_words"] = s.str.split().str.len()

    # Avoid division by zero for ratios
    df["n_words"] = df["n_words"].replace(0, np.nan)

    # Unique words
    df["n_unique_words"] = s.str.split().apply(
        lambda x: len(set(x)) if isinstance(x, list) and len(x) > 0 else 0
    )

    # Ratios
    df["ratio_unique_words"] = df["n_unique_words"] / df["n_words"]
    df["chars_per_word"] = df["n_chars"] / df["n_words"]

    # Fill NaNs back to 0 where appropriate
    df[["n_words", "ratio_unique_words", "chars_per_word"]] = df[
        ["n_words", "ratio_unique_words", "chars_per_word"]
    ].fillna(0.0)

    # ---------- Punctuation / special chars ----------
    df["n_exclamation"] = s.str.count("!")
    df["n_question"] = s.str.count(r"\?")
    df["n_period"] = s.str.count(r"\.")
    df["n_comma"] = s.str.count(",")

    # Hashtags / mentions (Twitter-style)
    df["n_hashtags"] = s.str.count(r"#\w+")
    df["n_mentions"] = s.str.count(r"@\w+")

    # URLs
    url_pattern = r"(https?://\S+|www\.\S+)"
    df["n_urls"] = s.str.count(url_pattern)

    # ---------- Capitalization ----------
    df["n_caps"] = s.apply(lambda x: sum(1 for ch in x if ch.isupper()))
    df["caps_ratio"] = df["n_caps"] / df["n_chars"].replace(0, np.nan)
    df["caps_ratio"] = df["caps_ratio"].fillna(0.0)

    # ---------- Simple lexical patterns ----------
    df["n_digits"] = s.str.count(r"\d")
    df["n_non_ascii"] = s.apply(lambda x: sum(ord(ch) > 127 for ch in x))

    # ---------- Whitespace patterns ----------
    df["n_newlines"] = s.str.count(r"\n")
    df["n_tabs"] = s.str.count(r"\t")

    # ---------- Normalize / clip ----------
    df["ratio_unique_words"] = df["ratio_unique_words"].clip(0, 1)

    # ---------- RoBERTa sentiment features ----------
    texts_list = s.tolist()
    if len(texts_list) > 0:
        sent_scores = compute_roberta_sentiment(texts_list, batch_size=32, max_len=128)
        # sent_scores shape: (N, 3) -> [neg, neu, pos]
        df["sent_neg_roberta"] = sent_scores[:, 0]
        df["sent_neu_roberta"] = sent_scores[:, 1]
        df["sent_pos_roberta"] = sent_scores[:, 2]
        df["sent_compound_roberta"] = df["sent_pos_roberta"] - df["sent_neg_roberta"]
    else:
        df["sent_neg_roberta"] = 0.0
        df["sent_neu_roberta"] = 0.0
        df["sent_pos_roberta"] = 0.0
        df["sent_compound_roberta"] = 0.0

    return df

# ============================================================
# 2. Apply to train / val / test with safe wrapper for val_df
# ============================================================

# Assumes df_train and val_df already exist and each has "text_for_model"

df_train_feats = add_handcrafted_features(df_train, text_col="text_for_model")

df_val_feats = None
if val_df is not None:
    df_val_feats = add_handcrafted_features(val_df, text_col="text_for_model")

# If you have test_df:
# df_test_feats = add_handcrafted_features(test_df, text_col="text_for_model")

# ============================================================
# 3. Quick check of created features
# ============================================================

print("Train handcrafted feature columns (new ones):")
train_new_cols = [c for c in df_train_feats.columns if c not in df_train.columns]
print(train_new_cols)

if df_val_feats is not None:
    print("\nVal handcrafted feature columns (new ones):")
    val_new_cols = [c for c in df_val_feats.columns if c not in val_df.columns]
    print(val_new_cols)
else:
    print("\nNo val_df provided (or val_df is None) → df_val_feats is None")

```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score, classification_report


def train_lightgbm_on_features(
    df_train_feats,
    df_val_feats=None,
    label_col="label",
    drop_cols=("text_for_model", "id"),
    label2id=None,
    num_classes=None,
):
    """
    df_train_feats: DataFrame with features + label
    df_val_feats:   DataFrame with features + label, or None
    label_col:      name of label column
    drop_cols:      columns to exclude from features if present
    label2id:       dict emotion -> int (optional, for naming)
    num_classes:    int, optional if label2id is not provided
    """

    # -----------------------------------
    # 1. Basic checks
    # -----------------------------------
    assert isinstance(df_train_feats, pd.DataFrame), "df_train_feats must be a DataFrame"
    if df_val_feats is not None:
        assert isinstance(df_val_feats, pd.DataFrame), "df_val_feats must be a DataFrame or None"

    # -----------------------------------
    # 2. Targets
    # -----------------------------------
    y_train = df_train_feats[label_col].values
    y_val = None
    if df_val_feats is not None and label_col in df_val_feats.columns:
        y_val = df_val_feats[label_col].values

    # Determine num_classes
    if num_classes is None:
        if label2id is not None:
            num_classes = len(label2id)
        else:
            num_classes = len(np.unique(y_train))

    # -----------------------------------
    # 3. Feature selection (numeric only)
    # -----------------------------------
    numeric_cols = [
        c for c in df_train_feats.columns
        if pd.api.types.is_numeric_dtype(df_train_feats[c])
    ]

    feature_cols = [
        c for c in numeric_cols
        if c != label_col and c not in drop_cols
    ]

    X_train = df_train_feats[feature_cols]
    X_val = None
    if df_val_feats is not None:
        X_val = df_val_feats[feature_cols]

    print("Using features:", feature_cols)
    print("X_train shape:", X_train.shape)
    if X_val is not None:
        print("X_val shape:", X_val.shape)
    else:
        print("No validation set provided; training without early stopping.")

    # -----------------------------------
    # 4. LightGBM model
    # -----------------------------------
    lgbm = lgb.LGBMClassifier(
        objective="multiclass",
        num_class=num_classes,
        learning_rate=0.05,
        n_estimators=2000,
        max_depth=-1,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        n_jobs=-1,
    )

    # -----------------------------------
    # 5. Fit (with or without val)
    # -----------------------------------
    if X_val is not None and y_val is not None:
        # With validation → use early stopping
        lgbm.fit(
            X_train,
            y_train,
            eval_set=[(X_val, y_val)],
            eval_metric="multi_logloss",
            early_stopping_rounds=50,
            verbose=50,
        )

        # Evaluate on val
        y_val_pred = lgbm.predict(X_val)
        val_acc = accuracy_score(y_val, y_val_pred)
        val_f1_macro = f1_score(y_val, y_val_pred, average="macro")

        print("\n=== LightGBM on features (with val) ===")
        print("Validation Accuracy:", val_acc)
        print("Validation Macro F1:", val_f1_macro)

        # Pretty names if you have id2label
        target_names = None
        if label2id is not None:
            id2label_local = {v: k for k, v in label2id.items()}
            target_names = [
                id2label_local.get(i, str(i)) for i in range(num_classes)
            ]

        if target_names is not None:
            print("\n=== Classification report (per class) ===")
            print(classification_report(y_val, y_val_pred, target_names=target_names))
        else:
            print("\n=== Classification report (per class, numeric labels) ===")
            print(classification_report(y_val, y_val_pred))
    else:
        # No validation set → just fit
        lgbm.fit(X_train, y_train)
        print("\n=== LightGBM on features (no val set) ===")
        print("Trained on full train set without early stopping.")

    # -----------------------------------
    # 6. Probabilities for stacking
    # -----------------------------------
    train_proba = lgbm.predict_proba(X_train)
    train_proba_cols = []
    if label2id is not None:
        id2label_local = {v: k for k, v in label2id.items()}
        class_names = [id2label_local.get(i, str(i)) for i in range(num_classes)]
    else:
        class_names = [str(i) for i in range(num_classes)]

    train_proba_cols = [f"lgbm_p_{name}" for name in class_names]
    train_proba_df = pd.DataFrame(train_proba, columns=train_proba_cols, index=df_train_feats.index)

    val_proba_df = None
    if X_val is not None:
        val_proba = lgbm.predict_proba(X_val)
        val_proba_df = pd.DataFrame(val_proba, columns=train_proba_cols, index=df_val_feats.index)

    return lgbm, train_proba_df, val_proba_df, feature_cols
```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import accuracy_score, f1_score, classification_report

import torch
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TextClassificationPipeline,
)

# ============================================================
# 1. Hand-crafted feature function
# ============================================================

def add_handcrafted_features(df, text_col="text_for_model"):
    df = df.copy()
    s = df[text_col].fillna("")

    # Basic length features
    df["len_chars"] = s.str.len()
    df["len_words"] = s.str.split().str.len()

    # Punctuation / intensity
    df["num_exclam"] = s.str.count("!")
    df["num_question"] = s.str.count(r"\?")

    # Uppercase ratio (shouting)
    def caps_ratio_fn(x):
        alpha = sum(c.isalpha() for c in x)
        if alpha == 0:
            return 0.0
        return sum(c.isupper() for c in x) / alpha

    df["caps_ratio"] = s.apply(caps_ratio_fn)

    # Structural: hashtags, mentions, urls
    df["num_hashtags"] = s.str.count("#")
    df["num_mentions"] = s.str.count("@")
    df["num_urls"] = s.str.count(r"http")

    return df

# ============================================================
# 2. Sentiment feature function (RoBERTa twitter sentiment)
# ============================================================

SENT_MODEL_NAME = "cardiffnlp/twitter-roberta-base-sentiment-latest"

sent_tokenizer = AutoTokenizer.from_pretrained(SENT_MODEL_NAME)
sent_model = AutoModelForSequenceClassification.from_pretrained(SENT_MODEL_NAME)
sent_model.eval()

# Use GPU if available, otherwise CPU
device = 0 if torch.cuda.is_available() else -1

sent_pipe = TextClassificationPipeline(
    model=sent_model,
    tokenizer=sent_tokenizer,
    top_k=None,       # same effect as return_all_scores=True
    device=device,
)

def get_sentiment_features(df, text_col="text_for_model", batch_size=64, max_len=64):
    df = df.copy()
    texts = df[text_col].fillna("").tolist()

    # Single call with internal batching (no manual for-loop over batches)
    outputs = sent_pipe(
        texts,
        batch_size=batch_size,
        truncation=True,
        max_length=max_len,
    )

    rows = []
    for out in outputs:  # out: list of dicts [{"label": "...", "score": ...}, ...]
        d = {}
        for item in out:
            label = item["label"].lower()  # "negative", "neutral", "positive"
            d[f"sent_{label}"] = item["score"]
        rows.append(d)

    sent_df = pd.DataFrame(rows, index=df.index)

    # Ensure all sentiment columns exist
    for col in ["sent_negative", "sent_neutral", "sent_positive"]:
        if col not in sent_df.columns:
            sent_df[col] = 0.0

    return sent_df

# ============================================================
# 3. Build feature DataFrames for train / val (if any) / test
# ============================================================

df_train_feats = add_handcrafted_features(df_train, text_col="text_for_model")
train_sent = get_sentiment_features(df_train_feats, text_col="text_for_model")
df_train_feats = pd.concat([df_train_feats, train_sent], axis=1)

HAS_VAL = "val_df" in globals() and val_df is not None

if HAS_VAL:
    df_val_feats = add_handcrafted_features(val_df, text_col="text_for_model")
    val_sent = get_sentiment_features(df_val_feats, text_col="text_for_model")
    df_val_feats = pd.concat([df_val_feats, val_sent], axis=1)
else:
    df_val_feats = None

df_test_feats = add_handcrafted_features(df_test, text_col="text_for_model")
test_sent = get_sentiment_features(df_test_feats, text_col="text_for_model")
df_test_feats = pd.concat([df_test_feats, test_sent], axis=1)

print(
    "Train feature columns added:",
    [c for c in df_train_feats.columns if c not in df_train.columns],
)

# ============================================================
# 4. Wrapper: train LightGBM with or without validation
# ============================================================

def train_lightgbm_on_features(
    df_train_feats,
    df_val_feats=None,
    label_col="label",
    drop_cols=("text_for_model", "id"),
    label2id=None,
    num_classes=None,
):
    assert isinstance(df_train_feats, pd.DataFrame), "df_train_feats must be a DataFrame"
    if df_val_feats is not None:
        assert isinstance(df_val_feats, pd.DataFrame), "df_val_feats must be a DataFrame or None"

    # Targets
    y_train = df_train_feats[label_col].values
    y_val = None
    if df_val_feats is not None and label_col in df_val_feats.columns:
        y_val = df_val_feats[label_col].values

    # Num classes
    if num_classes is None:
        if label2id is not None:
            num_classes = len(label2id)
        else:
            num_classes = len(np.unique(y_train))

    # Numeric feature selection
    numeric_cols = [
        c for c in df_train_feats.columns
        if pd.api.types.is_numeric_dtype(df_train_feats[c])
    ]
    feature_cols = [
        c for c in numeric_cols
        if c != label_col and c not in drop_cols
    ]

    X_train = df_train_feats[feature_cols]
    X_val = df_val_feats[feature_cols] if df_val_feats is not None else None

    print("Using features:", feature_cols)
    print("X_train shape:", X_train.shape)
    if X_val is not None:
        print("X_val shape:", X_val.shape)
    else:
        print("No validation set provided; training without early stopping.")

    # LightGBM model
    lgbm = lgb.LGBMClassifier(
        objective="multiclass",
        num_class=num_classes,
        learning_rate=0.05,
        n_estimators=2000,
        max_depth=-1,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1.0,
        n_jobs=-1,
    )

    # Fit
    if X_val is not None and y_val is not None:
        lgbm.fit(
            X_train,
            y_train,
            eval_set=[(X_val, y_val)],
            eval_metric="multi_logloss",
            early_stopping_rounds=50,
            verbose=50,
        )

        # Eval
        y_val_pred = lgbm.predict(X_val)
        val_acc = accuracy_score(y_val, y_val_pred)
        val_f1_macro = f1_score(y_val, y_val_pred, average="macro")

        print("\n=== LightGBM on features (with val) ===")
        print("Validation Accuracy:", val_acc)
        print("Validation Macro F1:", val_f1_macro)

        # Pretty report
        if label2id is not None:
            id2label_local = {v: k for k, v in label2id.items()}
            target_names = [id2label_local.get(i, str(i)) for i in range(num_classes)]
        else:
            target_names = [str(i) for i in range(num_classes)]

        print("\n=== Classification report (per class) ===")
        print(classification_report(y_val, y_val_pred, target_names=target_names))
    else:
        lgbm.fit(X_train, y_train)
        print("\n=== LightGBM on features (no val set) ===")
        print("Trained on full train set without early stopping.")

    return lgbm, feature_cols

# ============================================================
# 5. Train model
# ============================================================

if HAS_VAL:
    model_lgbm, feature_cols = train_lightgbm_on_features(
        df_train_feats,
        df_val_feats,
        label_col="label",
        drop_cols=("text_for_model", "id"),
        label2id=label2id,
    )
else:
    model_lgbm, feature_cols = train_lightgbm_on_features(
        df_train_feats,
        df_val_feats=None,
        label_col="label",
        drop_cols=("text_for_model", "id"),
        label2id=label2id,
    )


# ============================================================
# 6. Predict on test and build Kaggle submission
# ============================================================

X_test = df_test_feats[feature_cols]
test_pred = model_lgbm.predict(X_test)  # integer labels

# Map back to emotion strings
id2label_local = {v: k for k, v in label2id.items()}
test_emotion = [id2label_local[int(i)] for i in test_pred]

# Use post_id as the Kaggle "id"
if "post_id" in df_test_feats.columns:
    test_ids = df_test_feats["post_id"].values
elif "post_id" in df_test.columns:
    test_ids = df_test["post_id"].values
else:
    raise ValueError("No 'post_id' column found in df_test_feats or df_test.")

submission = pd.DataFrame({
    "id": test_ids,          # <- Kaggle needs this column name
    "emotion": test_emotion,
})

submission_path = "submission_lightgbm_feats_sentiment.csv"
submission.to_csv(submission_path, index=False)
print(f"\nSaved Kaggle submission to: {submission_path}")
```
